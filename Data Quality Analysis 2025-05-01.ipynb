### Срез по источникам данных

```python
df_stat = conn.sql("""
 
        SELECT
             source.uid as uid,
             source.name as name,
             source.url as url,
             SUM(CASE WHEN dataset.topics = [] THEN 1 END) as "Число датасетов без topics",
             COUNT(dataset.id) as "Число датасетов",
             (SUM(CASE WHEN dataset.topics = [] THEN 1 END))/(COUNT(dataset.id))*100 as "Доля без topics"
        FROM 'fulldb.parquet'
        GROUP BY 1, 2, 3
        ORDER BY 4 DESC;
    
    
""").df()

df_stat
```

### Срез по типу ПО каталога

```python
df_software = conn.sql(""" 
    SELECT
        source.software.name, 
        SUM(CASE WHEN dataset.topics = [] THEN 1 END) as "Число датасетов без topics",
        COUNT(dataset.id) as "Число датасетов",
        (SUM(CASE WHEN dataset.topics = [] THEN 1 END))/(COUNT(dataset.id))*100 as "Доля без topics"
    FROM 'fulldb.parquet' 
    GROUP BY 1
    ORDER BY 4 DESC;
""").df()

df_software
```

```python
df_toobig_desc_topics = conn.sql("""
    SELECT 
        source.uid,
        source.name as source_name,
        dataset.description,
        COUNT(dataset.description),
        LENGTH(dataset.description),
        SUM(CASE WHEN dataset.topics = [] THEN 1 ELSE 0 END) as "Число датасетов без topics",
        COUNT(dataset.id) as "Число датасетов",
        (SUM(CASE WHEN dataset.topics = [] THEN 1 ELSE 0 END))/(COUNT(dataset.id))*100 as "Доля без topics"
    FROM 'fulldb.parquet'
    WHERE LENGTH(dataset.description) > 1000
    GROUP BY 1, 2, 3
    HAVING COUNT(dataset.description)>1
    ORDER BY 7 DESC, 8 DESC;
""").df()

df_toobig_desc_topics
```

### Ищем несодержательные плохие описания
Первое приближение. Использование регулярных выражений.

Напишем проверку с помощью регулярного выражения: ищем описания, где один символ повторяется 3 и более раз. Символы могут быть буквенные, числовые и иные специальные.

Дисклеймер 1: числа (денежные значения) могут иметь повторяющиеся символы, например, 0, поэтому этот способ не идеален, но можно получить выборку для анализа и использования в повышении качества описаний.

Дисклеймер 2: не получается продемонстрировать использование регулярного выражения на таблице, где есть поля source.uid, source.name, source.url -- Out of Memory. Эти поля полезны для знания, где искать подобные датасеты

```python
#вытащим все уникальные описания датасетов и посчитаем их количество
df_s = conn.sql(""" 
        SELECT 
            -- source.uid,
            -- source.name,
            -- source.url,
            dataset.description,
            LENGTH(dataset.description),
            COUNT(dataset.id)
        FROM 'fulldb.parquet'
        WHERE dataset.description IS NOT NULL
        GROUP BY 1
        ORDER BY 3 DESC, 2 DESC;
""").df()

df_s
```

Используем регулярное выражение **r'((.)\2{3,})'** и маску для фильтрации датафрема -- остались описания с повторяющимися символами. 

```python
# 3 и более повторяющихся символа
mask = df_s.description.str.match(r'((.)\2{3,})')
df_s[mask]
```

## Разметка языка датасетов

Источники (source) имеют размеченный язык в поле **source.langs**. 
Что если источник может быть англоязычным, а датасет на другом языке?

Используя библиотеку **langdetect**, проверим на **100 тыс.** описаний датасетов. 

```python
#оставляю 1 млн описаний датасетов
df_cut = df_s[:100000]
```

```python
import pandas as pd
from langdetect import detect, DetectorFactory

DetectorFactory.seed = 0


# Функция для определения языка
def detect_language(text):
    try:
        return detect(text)
    except:
        return 'unknown'

# Создаём новый столбец с кодом языка
df_cut['dataset.language'] = df_cut['description'].apply(detect_language)

# Просмотр результата
print(df_cut[['description', 'dataset.language']].head())
```

```python
#просмотр таблицы с размеченным языком для каждого описания
df_cut
```

**langdetected** разметила описания двубуквенным значением, например, "en", используя стандарт ISO 639, добавим расшифровку языка.

```python
import pandas as pd

#чтение файла с расшифровкой кодов языка
df_excel = pd.read_excel('List_of_ISO_639_language_codes.xlsx') 

#объединение таблиц по двубуквенному коду языка
df_merged = df_cut.merge(df_excel, how='left', on='dataset.language')

#просмотр таблицы
df_merged
```

Просмотрим, какие языки встречаются чаще всего.

```python
grouped_counts = df_merged.groupby('ISO Language Names').size().sort_values(ascending=False)
grouped_counts
```

Сравним с результатом, который получится, по 100 тыс. случайным датасетом, язык который размечен по источнику.

```python
df_language = conn.sql("""
    SELECT 
        source.name,
        source.url,
        (source.langs::json->0->>'name') as lang,
        dataset.description,
        COUNT(dataset.id) as "num_datasets",
    FROM 'fulldb.parquet'
    WHERE LENGTH(dataset.description) >=1
    GROUP BY 1, 2, 3, 4
    ORDER BY 5 DESC
    LIMIT 100000;
""").df()

df_language
```

```python
df_langsource = df_language.groupby('lang').size().sort_values(ascending=False)
df_langsource
```

В целом, так как использовались 100 тыс. случайных датасетов, можно сказать, что результаты по языку сопоставимы с распознаванием языка по описаниям с помощью langdetect.

Посмотрим, каким языком размечены источники во всей таблицы.

```python
df_language = conn.sql("""
    SELECT 
        source.name,
        source.url,
        (source.langs::json->0->>'name') as lang,
        COUNT(dataset.id) as "num_datasets"
    FROM 'fulldb.parquet'
    WHERE LENGTH(dataset.description) >=1
    GROUP BY 1, 2, 3
    ORDER BY 4 DESC;
""").df()

df_language
```

## Срезы данных по форматам, типам данных, периодам и частоте обновлений

Дисклеймер: метод *unnest(dataset.formats/datatypes)* -- не дает отработать на всем массиве данных, поэтому тут стоит ограничение **1 млн**. 

Срез по форматам данных

```python
df_formats = conn.sql("""
    WITH t1 AS 
        (SELECT
            unnest(dataset.formats) as elem
        FROM 'fulldb.parquet'
        LIMIT 1000000)
    
    SELECT 
        elem,
        COUNT(*) as "Число датасетов"
    FROM t1
    GROUP BY 1
    ORDER BY 2 DESC;
""").df()

df_formats
```

Срез по типам данных

```python
df_datatypes = conn.sql("""
    WITH t1 AS 
        (SELECT
            unnest(dataset.datatypes) as elem
        FROM 'fulldb.parquet'
        LIMIT 1000000)
    
    SELECT 
        elem,
        COUNT(*) as "Число датасетов"
    FROM t1
    GROUP BY 1
    ORDER BY 2 DESC;
""").df()

df_datatypes
```

Срез по частоте обновлений

```python
df_frequency = conn.sql("""
    SELECT
        dataset.frequency as elem,
        COUNT(*)
    FROM 'fulldb.parquet'
    GROUP BY 1
    ORDER BY 2 DESC;
    
""").df()

df_frequency
```

**18,6 млн** или **84,27%** датасетов не имеют значения в поле **frequency** - частота обновления

Срез по периодам (временным рядам)

```python
df_years = conn.sql("""
    SELECT
        dataset.year_range as elem,
        COUNT(*)
    FROM 'fulldb.parquet'
    GROUP BY 1
    ORDER BY 2 DESC;
    
""").df()

df_years
```

Точное совпадение по числу датасетов, не имеющих узказанного временного периода: те же **18,6 млн** или **84,27%** датасетов не имеют значения в поле **year_range**.
Полагаю, что есть прямая связь между заполнением временного периода и указания частоты обновления датасета.

```python
# Разворачиваем начало и конец временного периода в отдельные колонки
elem_expanded = df_years['elem'].apply(pd.Series)

# Если в значениях есть кавычки, уберём их
elem_expanded = elem_expanded.applymap(lambda x: x.strip('"') if isinstance(x, str) else x)

# Объединяем с исходным DataFrame, сохраняя колонку с количеством
result = pd.concat([elem_expanded, df_years['count_star()']], axis=1)

print(result)
```

Срез по началу временного периода для датасетов, где поле year_range заполнено

```python
df_start = result.groupby('start')['count_star()'].sum().sort_values(ascending=False)
df_start
```

Срез по концу временного периода для датасетов, где поле year_range заполнено

```python
df_end = result.groupby('end')['count_star()'].sum().sort_values(ascending=False)
df_end
```
