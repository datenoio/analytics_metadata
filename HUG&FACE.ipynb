### Проверка на полноту метаданных

Проверим, существуют ли датасеты без заголовков.

```python
df_titlenan = conn.sql("""
SELECT * 
FROM hf_table
WHERE title IS NULL;
        """). df()

df_titlenan
```

Теперь посмотрим на заполненность описаний (description и short_text).

```python
df_descnan = conn.sql("""
SELECT 
    COUNT(id)
FROM hf_table
WHERE description IS NULL;
        """). df()
df_descnan
```

```python
df_shortnan = conn.sql("""
SELECT 
    COUNT(id)
FROM hf_table
WHERE short_text IS NULL;
        """). df()
df_shortnan
```

**247 057** датасетов не имеют никаких описаний. Это **68,02%**. Тут однозначно есть зависимость между заполненностью **description** и **short_text**. 



Проверим заполненность поля **topics**.
Не указано **topics** для **343 147** или **94.48%** не имеют указанных topics.
Напомню, что доля заполненности описаний выше.
То есть **topics** можно обогатить на основе этого поля и повысить заполненность до **32%**. 

```python
df_topics = conn.sql("""
SELECT 
    topics,
    COUNT(id)
FROM hf_table
GROUP BY 1
ORDER BY 2 DESC;
        """). df()
df_topics
```

Заполненность **geotopics** хуже. Значения отсутствуют у **348 407** или **95.93%** датасетов.

```python
df_geotopics = conn.sql("""
SELECT 
    geotopics,
    COUNT(id)
FROM hf_table
GROUP BY 1
ORDER BY 2 DESC;
        """). df()
df_geotopics
```

Посмотрим на заполненность поля **license**. 
**290 025** или **81,2%** не имеют указанной лицензии. 
Для остальных датасетов лидируют 
- Apache License 2.0 (**28 227 или 7.7%**)
- MIT License (**26 842 или 7.39%**)
- Creative Commons 4.0 (**5 231 или 1.44%**)

```python
df_license = conn.sql("""
SELECT 
    (license::json->>'name') as license,
    COUNT(id)
FROM hf_table
GROUP BY 1
ORDER BY 2 DESC;
        """). df()
df_license
```

```python
df_license_topics = conn.sql(""" 
    SELECT
        (license::json->>'name') as license, 
        SUM(CASE WHEN topics IS NULL THEN 1 END) as "Число датасетов без topics",
        COUNT(id) as "Число датасетов",
        (SUM(CASE WHEN topics IS NULL THEN 1 END))/(COUNT(id))*100 as "Доля незаполненности"
    FROM hf_table
    GROUP BY 1
    ORDER BY 4 DESC;
""").df()

df_license_topics
```

Какие регионы и страны указываются при заполнении датасетов?
У **363 118** или **99.98%** датасетов указаны США.

```python
df_regions = conn.sql("""
SELECT 
    (regions::json->0->>'region_name') as macroregions,
    (regions::json->0->>'country_or_area') as country,     
    COUNT(id)
FROM hf_table
GROUP BY 1, 2
ORDER BY 3 DESC;
        """). df()
df_regions
```

```python
df_country_topics = conn.sql(""" 
    SELECT
        (regions::json->0->>'region_name') as macroregions,
        (regions::json->0->>'country_or_area') as country, 
        SUM(CASE WHEN topics IS NULL THEN 1 END) as "Число датасетов без topics",
        COUNT(id) as "Число датасетов",
        (SUM(CASE WHEN topics IS NULL THEN 1 END))/(COUNT(id))*100 as "Доля незаполненности"
    FROM hf_table
    GROUP BY 1, 2
    ORDER BY 5 DESC;
""").df()

df_country_topics
```
